```python
import pandas as pd
df = pd.read_csv(r"C:\Users\ASUS\Documents\GitHub\Data-Science-Book\GroupHCMUS.csv")
```

```python
# Xóa đoạn thừa phía trước "Chia sẻ với Nhóm công khai"
df["Post"] = df["Post"].str.split("công khai", n=1).str[1]
df.dropna(subset="Post", inplace = True)
# Xác định dòng chứa tên tài khoản cá nhân
mask = df["Post"].str.contains("youraccountname")
# Sửa cột post chỉ giữ lại phần phía sau tên tài khoản cá nhân
df.loc[mask, "Post"] = df.loc[mask, "Post"].str.split("youraccountname", n=1).str[1]
df
```

```python
# Tách từ dữ liệu gốc thành hai cột "Post" chứa nội dung bài viết và "React" chứa thông tin về lượt tương tác
df[["Post", "React"]] = df["Post"].str.split("Tất cả cảm xúc:", n=1, expand=True)
df["React"] = df["React"].str.split(" ").str[0]
df
```

```python
# Xóa các bài viết không có react
df.dropna(subset="React", inplace=True)
```

```python
mask = df["Post"].str.contains("Đã chia sẻ với Công khai")
df.loc[mask, "Bài Share lại"] = True
df.loc[mask, "Post"] = df.loc[mask, "Post"].str.split("Đã chia sẻ với Công khai", n=1).str[1]
df["Bài Share lại"] = df["Bài Share lại"].fillna(False)
```

```python
df.to_csv("yourcsv.csv")
```
